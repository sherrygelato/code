```python
#-*- coding: utf-8 -*-
from keras.models import Sequential # 순차적으로 쌓는데
from keras.layers import Dense # Dense로 쌓을거야
from sklearn.model_selection import train_test_split # set 나눌거야

import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
# seed 값 설정
seed = 0
np.random.seed(seed)
tf.random.set_seed(3) # 동일한 결과 나오도록 

df = pd.read_csv("/content/drive/MyDrive/data/housing.csv", delim_whitespace=True, header=None)
# delim_whitespace 한줄씩 띄어쓰기 함 (구분절)
```

```python
print(df.info())
print()
print(df.head())
```

```python
# 레코드는 다 가져오고 속성 정보는 나눠줌
# 분류문제가 아니라 연속형 변수
dataset = df.values
X = dataset[:,0:13]
Y = dataset[:,13]
```

```python
# 데이터 분리
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=seed)
```

```python
# 모델 생명
model = Sequential()
model.add(Dense(30, input_dim=13, activation='relu')) # 13개 독립 변수 받고 그대로 relu로 나가고
model.add(Dense(6, activation='relu'))
model.add(Dense(1)) # 회귀 문제이기 때문에 활성화 함수가 없다
```

```python
# 환경 설정
model.compile(loss='mean_squared_error', # 평균제곱오차
              optimizer='adam')
```

```python
# 훈련용 데이터로 모델 훈련
model.fit(X_train, Y_train, epochs=200, batch_size=10)
```

```python
# 분류 문제가 아니라 예측 문제다 보니
# 예측 값과 실제 값의 비교
Y_prediction = model.predict(X_test).flatten() # n차원이든 1차원으로 만들어줌 flatten()
# 1차원 배열로 해야 컴퓨터가 이해 함
for i in range(10):
    label = Y_test[i]
    prediction = Y_prediction[i]
    print("실제가격: {:.3f}, 예상가격: {:.3f}".format(label, prediction))
```
