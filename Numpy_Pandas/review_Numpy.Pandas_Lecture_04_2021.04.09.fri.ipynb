## 2. Matplotlib 그래프

- 컬러 https://matplotlib.org/stable/tutorials/colors/colors.html

- Line plot

```python
fig, ax = plt.subplots()

x = np.arange(15)
y = x**2

ax.plot(
    x,y,
    linestyle = ":",
    marker = "*",
    color = "#1f77b4"
)
```

- Line style 

```python
x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(x, x, linestyle="-")
# solid
ax.plot(x, x+2, linestyle="--")
# dashed
ax.plot(x, x+4, linestyle="-.")
# dashdot
ax.plot(x, x+6, linestyle=":")
# dotted
```

```python
# Marker

x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(x, x, marker=".")
ax.plot(x, x+2, marker="o") # 큰 원
ax.plot(x, x+4, marker='v') # 아래 삼각형
ax.plot(x, x+6, marker="s") # 사각형(s: square)
ax.plot(x, x+8, marker="*")
```

- 축 경계 조정

```python
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x))
ax.set_xlim(-2, 12)     # xlim : x 축 경계
ax.set_ylim(-1.5, 1.5)
```

- 범례 

```python
fig, ax = plt.subplots()
ax.plot(x, x, label='y=x')
ax.plot(x, x**2, label='y=x^2')
ax.set_xlabel("x")
ax.set_ylabel("y")

# 범례 데이터 
ax.legend( # 범례: legend
    loc='upper right',  # loc : 범례위치
    shadow=True,       # shadow = True 그림자 효과
    fancybox=True,     # fancybox = True 모서리를 둥글게 만든 박스
    borderpad=2)       # borderpad =2     흰색 크기 조정
```

```python
fig, ax = plt.subplots()
ax.plot(x, x, label='y=x')
ax.plot(x, x**2, label='y=x^2')
ax.set_xlabel("x")
ax.set_ylabel("y")

# 범례 데이터 
ax.legend( # 범례: legend
    loc='upper left',  # loc : 범례위치
    shadow=True,       # shadow = True 그림자 효과, 범례의 그림자야
    fancybox=True,     # fancybox = True 모서리를 둥글게 만든 박스
    borderpad=2)       # borderpad =2     흰색 크기 조정
```

```python
x = np.arange(10)
fig, ax = plt.subplots() ###### 외우세요
ax.plot(
    x, x, label='y=x', # 처음x는 나중 x는 y의 값.
    marker='o',
    color='blue',
    linestyle=':'
)
ax.plot(
    x, x**2, label='y=x^2',
    marker='^',
    color='red',
    linestyle='--'
)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend(
    loc='upper left',
    shadow=True,
    fancybox=True,
    borderpad=2
)

fig.savefig("plot.png") # 이 그림 저장해주세요.
```

## 3. Scatter

```python
fig, ax = plt.subplots()
x = np.arange(10)
ax.plot(
  x, x**2, 'o', # 'o'로 인해서 산점도 그래프로 변한다(세번째 인자)
  markersize = 15,
  markerfacecolor = 'white',
  markeredgecolor = "blue"
)
```

```python
fig, ax = plt.subplots()
x = np.random.randn(50)
y = np.random.randn(50)
colors = np.random.randint(0, 100, 50)
sizes = 500 * np.pi * np.random.rand(50) ** 2
ax.scatter(x, y, c=colors, s=sizes, alpha=0.3)  #알파는 투명도
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
x = np.arange(10)
ax.plot(
    x, x**2, "o",
    markersize=15,
    markerfacecolor='white',
    markeredgecolor="blue"
)

fig.savefig("plot.png")
```

## 4. Bar & Histogram

```python
x = np.arange(10)
fig, ax = plt.subplots(figsize=(12, 4))
ax.bar(x, x**2)

fig.savefig("plot.png") # 이미지 파일엔 그래프 나옴
```

```python
x = np.random.rand(3)
y = np.random.rand(3)
z = np.random.rand(3)
data = [x, y, z]

fig, ax = plt.subplots()
x_ax = np.arange(3)

for i in x_ax:
    ax.bar(x_ax, data[i],
           bottom=np.sum(data[:i], axis=0))
    
ax.set_xticks(x_ax)
ax.set_xticklabels(["A", "B", "C"])
```

- histogram

```python
fig, ax = plt.subplots()
data = np.random.randn(1000)
ax.hist(data, bins=50)

fig.savefig("plot.png") ######### 확률
```
